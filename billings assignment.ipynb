```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\VIRAJ\Downloads\Dataset.csv",header=3)
df
```

```python
df.dropna(axis=1,how="all",inplace=True)
df
```

```python
df.loc[1]=df.loc[1].ffill()
```

```python
df.head()
```

```python
df.loc[2]=df.loc[0]+"/"+df.loc[1]+"/"+df.loc[2]
```

```python
df
```

```python
frow=df.loc[2]
frow[0]="date"
df.columns=frow
df=df[3:]
df=df.rename_axis(None,axis=1)
df
```

```python
l1=df.columns.tolist()[1:]
df=df.melt(id_vars=["date"],value_vars=l1)
df
```

```python
df["variable"]=df["variable"].str.replace("-","/")
df[["segment","period","type","subtype"]]=df["variable"].str.split("/",expand=True)
df.drop("variable",axis=1,inplace=True)
df=df[["segment","period","subtype","value","date"]]
df
```
